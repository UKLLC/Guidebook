```python
# for >1  dataset version
def get_latest_dsv_activity(source_table):
    r1 = requests.get(
            "https://api.test.datacite.org/dois?prefix=10.83126&query=titles.title:%22{}%22".format(source_table),
            headers={
                "accept": "application/vnd.api+json",
                    },
            auth=(ID, pw)
            )

    doi_df = pd.json_normalize(r1.json()["data"])
    doi_vn = doi_df[doi_df["attributes.version"] == doi_df["attributes.version"].max()].iloc[0]['id']

    r2 = requests.get(
            "https://api.test.datacite.org/dois/{}/activities".format(doi_vn),
            headers={
                "accept": "application/vnd.api+json",
                    },
            auth=(ID, pw)
            )

    return pd.json_normalize(r2.json()["data"])
```

```python
dfa = get_latest_dsv_activity("BCS70_bcs10_housing")
```

```python
dfa2 = dfa.iloc[2].dropna()
ch = dfa2[dfa2.index.str.startswith('attributes.changes')]
ch0 = ch.map(lambda x: x[0]).to_frame().rename(columns={2:"old"})
ch1 = ch.map(lambda x: x[1]).to_frame().rename(columns={2:"new"})
ch_both = ch0.join(ch1).T
ch_both["main title"] = ch_both["attributes.changes.titles"].apply(lambda x: x[0]["title"])
ch_both["alt_title"] = ch_both["attributes.changes.titles"].apply(lambda x: x[1]["title"])
ch_both = ch_both.loc[:, ~ch_both.eq(ch_both.iloc[0]).all()].drop("attributes.changes.titles", axis=1)
ch_both
```

```python
dfa4 = dfa.iloc[4].dropna()
ch = dfa4[dfa4.index.str.startswith('attributes.changes')]
ch0 = ch.map(lambda x: x[0]).to_frame().rename(columns={4:"old"})
ch1 = ch.map(lambda x: x[1]).to_frame().rename(columns={4:"new"})
ch_both = ch0.join(ch1).T
ch_both
```
