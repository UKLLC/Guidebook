```python
import requests
import json
import os
import sqlalchemy
import pandas as pd
import mdapi_functions as md
from IPython.display import display, Markdown
import markdown
from bokeh.plotting import figure, show
from bokeh.models import Span
from bokeh.models import TabPanel, Tabs, Tooltip
from bokeh.io import output_notebook
from math import pi
API_KEY = os.environ['FASTAPI_KEY']
```

```python
def get_md_api_frz_link_nhse():
    r_d = requests.get(
            "https://metadata-api-4a09f2833a54.herokuapp.com/freeze-linkage-nhse/",
            headers={'access_token': API_KEY.strip()})

        # return error message if API response != 200
    try:
        r_d.raise_for_status()

    except requests.exceptions.HTTPError as err:
        return "HTTP Error: " + str(err)

    return pd.DataFrame(json.loads(r_d.text))
```

```python
dff = get_md_api_frz_link_nhse()
dff = dff[(dff["LPS"] == "ALSPAC") & (dff["frz_num"] == dff["frz_num"].max())]

lx = ["NHS England", "NHS Wales", "NHS Scotland", "Neighbourhood Geographies", "Address Geographies", "DfE", "DWP", "HMRC"]
ly = [dff.iloc[0]["n_l_tot"], 0, 0, 0, 0, 0, 0, 0]

output_notebook(hide_banner=True)
p = figure(x_range=lx, width=600, height=400, title="Linkage of {} participants to linked Datasets".format("ALSPAC"))
p.vbar(x=lx, top=ly, width=0.8)
vline = Span(location=dff.iloc[0]["n_sent"], dimension='width', line_color='grey', line_width=2)
p.add_layout(vline)
p.y_range.start = 0
p.xaxis.major_label_orientation = pi/4

show(p)
```

```python
output_notebook(hide_banner=True)
dff = get_md_api_frz_link_nhse()

dff1 = dff[(dff["LPS"] == "ALSPAC") & (dff["frz_num"] == 1)]
dff2 = dff[(dff["LPS"] == "ALSPAC") & (dff["frz_num"] == 2)]
dff3 = dff[(dff["LPS"] == "ALSPAC") & (dff["frz_num"] == 3)]

ly1 = [dff1.iloc[0]["n_l_tot"], 0, 0, 0, 0, 0, 0, 0]
ly2 = [dff2.iloc[0]["n_l_tot"], 0, 0, 0, 0, 0, 0, 0]
ly3 = [dff3.iloc[0]["n_l_tot"], 0, 0, 0, 0, 0, 0, 0]

lx = ["NHS England", "NHS Wales", "NHS Scotland", "Neighbourhood Geographies", "Address Geographies", "DfE", "DWP", "HMRC"]

p1 = figure(x_range=lx, width=600, height=400, title="Linkage of {} participants to linked Datasets".format("ALSPAC"))
p1.vbar(x=lx, top=ly1, width=0.8)
vline1 = Span(location=dff1.iloc[0]["n_sent"], dimension='width', line_color='grey', line_width=2)
p1.add_layout(vline1)
p1.y_range.start = 0
p1.xaxis.major_label_orientation = pi/4

p2 = figure(x_range=lx, width=600, height=400, title="Linkage of {} participants to linked Datasets".format("ALSPAC"))
p2.vbar(x=lx, top=ly2, width=0.8)
vline2 = Span(location=dff2.iloc[0]["n_sent"], dimension='width', line_color='grey', line_width=2)
p2.add_layout(vline2)
p2.y_range.start = 0
p2.xaxis.major_label_orientation = pi/4

p3 = figure(x_range=lx, width=600, height=400, title="Linkage of {} participants to linked Datasets".format("ALSPAC"))
p3.vbar(x=lx, top=ly3, width=0.8)
vline3 = Span(location=dff3.iloc[0]["n_sent"], dimension='width', line_color='grey', line_width=2)
p3.add_layout(vline3)
p3.y_range.start = 0
p3.xaxis.major_label_orientation = pi/4

tabs = Tabs(tabs=[
    TabPanel(child=p1, title="Fz 1"),
    TabPanel(child=p2, title="Fz 2"),
    TabPanel(child=p3, title="Fz 3")
], active=2, height=400)

show(tabs)
```

```python
plist = []
lx = ["NHS England", "NHS Wales", "NHS Scotland", "Neighbourhood Geographies", "Address Geographies", "DfE", "DWP", "HMRC"]

for i in range(1,dff["frz_num"].max()+1):
    dffi = dff[(dff["LPS"] == "ALSPAC") & (dff["frz_num"] == i)]
    ly = [dffi.iloc[0]["n_l_tot"], 0, 0, 0, 0, 0, 0, 0]
    p = figure(x_range=lx, width=600, height=400, title="Linkage of {} participants to linked Datasets".format("ALSPAC"))
    p.vbar(x=lx, top=ly, width=0.8)
    vline = Span(location=dffi.iloc[0]["n_sent"], dimension='width', line_color='grey', line_width=2)
    p.add_layout(vline)
    p.y_range.start = 0
    p.y_range.end = round(dff[(dff["LPS"] == "ALSPAC")]["n_sent"].max()*1.2, -3)
    p.xaxis.major_label_orientation = pi/4
    plist.append(p)

tabs = Tabs(tabs=[
    TabPanel(child=plist[x-1], title="Freeze {}".format(x)) for x in range(1,dff["frz_num"].max()+1)
], active=dff["frz_num"].max()-1)

show(tabs)
```
