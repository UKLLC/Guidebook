## Townsend Index 

The Townsend index [(1989)](https://doi.org/10.4324/9781003368885) is a relative indicator of material deprivation, which has been widely used to identify socio‚Äêeconomic confounding in crime, education, and health research in England. This dataset is a **2011** version of the Townsend Index, created at the [UK Data Service](http://dx.doi.org/10.5257/census/aggregate-2011-2).
